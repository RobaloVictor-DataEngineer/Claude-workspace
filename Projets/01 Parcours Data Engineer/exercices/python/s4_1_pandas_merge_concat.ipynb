# Combiner des DataFrames : merge & concat (cours)

Exemples à **exécuter** (contexte : commandes + clients). Les exercices (commandes web) sont tout en bas.

### Deux tables à joindre

```python
import pandas as pd
commandes = pd.DataFrame({"id_commande":[1,2,3],"id_client":[10,11,10],"montant":[100,250,80]})
clients   = pd.DataFrame({"id_client":[10,11,12],"nom":["Alice","Bob","Chloé"],"ville":["Rouen","Paris","Lyon"]})
display(commandes)
display(clients)
```

### merge (le JOIN de pandas) + les how

```python
print("inner (défaut) :")
display(pd.merge(commandes, clients, on="id_client"))
print("left (garde toutes les commandes) :") 
display(pd.merge(clients, commandes, on="id_client", how="left"))
```

### Piège : inner peut perdre des lignes -> comparer les tailles

```python
fusion = pd.merge(commandes, clients, on="id_client")   # inner
print("commandes de départ :", len(commandes), "| après jointure :", len(fusion))
```

### concat : empiler

```python
jan = pd.DataFrame({"produit":["A","B"],"qte":[10,5]})
fev = pd.DataFrame({"produit":["A","C"],"qte":[7,3]})
print(pd.concat([jan, fev], ignore_index=True))
```

# Exercices

**Trois jeux de données, chargés dans la cellule ci-dessous :**
- `commandes_web` — les commandes du **mois 1** (une ligne = une commande, avec `client_id`, `montant`).
- `clients_web` — le **catalogue des clients connus** (`client_id`, nom, `segment`, `ville`).

**À toi de choisir tes outils.**

2. Combien de commandes dans `commandes_web` au départ ? Compare avec la question 1 : la jointure a-t-elle **perdu des commandes** ? Pourquoi ? Refais-la pour **garder les 60 commandes du mois 1**, même les « invité » (`client_id` absent de `clients_web`).
3. Sur ce tableau complet, calcule le **CA par segment**. Que fais-tu des commandes « invité » sans segment ?
4. Toujours sur ce tableau complet, calcule le **CA par ville**, du plus élevé au plus faible.
5. **Sans le catalogue clients** : mets bout à bout `commandes_web` (mois 1) et `commandes_web_mois2` (mois 2), puis calcule le **CA total des deux mois**. (empiler n'est pas une jointure)
6. Réflexion (pas de code) : différence entre une jointure `inner` et une jointure `left`, et quand choisir l'une plutôt que l'autre ?

```python
# --- CELLULE FOURNIE : données chargées pour toi ---
import pandas as pd
commandes_web       = pd.read_csv("data/commandes_web.csv")       # mois 1
clients_web         = pd.read_csv("data/clients_web.csv")          # catalogue clients
commandes_web_mois2 = pd.read_csv("data/commandes_web_mois2.csv")  # mois 2 (question 5 uniquement)
print("commandes_web (mois 1):", commandes_web.shape, "| clients_web:", clients_web.shape, "| mois2:", commandes_web_mois2.shape)
```

```python
## 1 & 2
print("Affichage de la table client :")
display(clients_web, len(clients_web))
print("Affichage de la table commandes :")
display(commandes_web, len(commandes_web))

# Avec INNER JOIN
print("Affichage du merge de commandes et clients (INNER) :")
df_inner = pd.merge(commandes_web, clients_web, on="client_id")
print("Nombre de lignes dans la table (INNER) : ", len(df_inner))
display(df_inner)

# Avec LEFT JOIN
print("Affichage du merge de commandes et clients (LEFT) :")
df_left = pd.merge(commandes_web, clients_web, on="client_id", how="left")
print("Nombre de lignes dans la table (LEFT) : ", len(df_left))
display(df_left)

# On a 15 lignes dans clients et 60 dans commandes (mois 1).
# D'après l'affichage on perd 13 lignes et ce qui est normal car on a dans commandes des client_id > 15 donc pas de correspondances, donc n'apparaissent pas
# Avec un LEFT on ne perd rien mais les clients_id qui ne sont pas dans clients auront des valeurs NaN pour nom, segment et ville
```

```python
## 3
CA_segment = df_left.groupby("segment")["montant"].sum().sort_values(ascending=False).reset_index()
display(CA_segment)


CA_segment_2 = df_left.fillna("Inconnu").groupby("segment")["montant"].sum().sort_values(ascending=False).reset_index()
display(CA_segment_2)

# Comme on peut le voir dans le cas 1, je n'ai pas précisé dropna=False donc le groupby va drop les valeur NaN automatiquement.
# Si on laisse le dropna=false comme dans le cas 2 alors on aura aussi un segment NaN compris dans le df affiché.
# Comme pour pleins de cas ça dépend de ce qu'on veut en entreprise si ce qui nous intéresse c'est le CA total et les clients qui ne sont pas dans client_id sont quand même valides alors on garde la solution #2
# mais si on  veut seulement les CA des segments valides alors on prends la solution #1
```

```python
## 4
CA_ville = df_left.groupby("ville")["montant"].sum().sort_values(ascending=False)
display(CA_ville.to_frame(name="CA total").reset_index())

CA_ville_2 = df_left.groupby("ville", dropna=False)["montant"].sum().sort_values(ascending=False)
display(CA_ville_2.to_frame(name="CA total").reset_index())

#Même raisonnement qu'à la question précédente
```

```python
## 5
print("Mois 1 de commandes : ")
display(commandes_web)

print("Mois 2 de commandes : ")
display(commandes_web_mois2)

all_commandes = pd.concat([commandes_web, commandes_web_mois2], ignore_index=True)
print("Mois 1 & 2 de commandes : ")
display(all_commandes)

# Calcul CA total :

ca_total = all_commandes["montant"].sum()
print(f"Le CA total sur 2 mois est de {ca_total}€")
```

```python
## 6 (réflexion en commentaire)
# INNER JOIN : Garde les éléments qui correspondent dans les 2 tables si un id n'existe pas dans une table et qu'on inner join sur cet id alors il n'apparaitra pas dans la table jointe
# LEFT JOIN : Garde tous les éléments de la table gauche et remplace par NaN si pas de correspondance dans la table de droite
# Exemple : j'ai une table avec des clients et des commandes (liés par un client_id), si je veux les commandes passés par les clients enregistrés uniquement je fais un INNER JOIN avec client je vais avoir uniquement les commandes ou client_id de commandes = client_id de clients
# Mais si je fais un LEFT JOIN sur client je vais avoir toutes les commandes mais les données reliés aux clients (qui ne sont pas dans la table client mais qui sont dans commandes) seront des NaN
```
