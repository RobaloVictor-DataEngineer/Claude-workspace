```python
employes = [
    {"prenom": "Victor", "salaire": 42000, "ville": "Rouen"},
    {"prenom": "Sarah",  "salaire": 38000, "ville": "Paris"},
    {"prenom": "Karim",  "salaire": 47000, "ville": "Rouen"},
    {"prenom": "Lina",   "salaire": 51000, "ville": "Lyon"},
]
```

```python
prenoms_employes = [n["prenom"] for n in employes]
print("1", prenoms_employes)
salaire_augmente = [s["salaire"] * 1.05 for s in employes]
print("2", salaire_augmente)
prenoms_employes_rouen = [nr["prenom"] for nr in employes if nr["ville"] == "Rouen"]
print("3", prenoms_employes_rouen)
prenoms_salaire = {p["prenom"]:p["salaire"] for p in employes}
print("4", prenoms_salaire)
prenoms_employes_plus_40k = [nr["prenom"] for nr in employes if nr["salaire"] > 40000]
print("5", prenoms_employes_plus_40k)


```
