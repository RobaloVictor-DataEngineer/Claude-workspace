# Pandas — sélection & filtrage avancés (cours)

Cellules d'exemple à **exécuter** (contexte : table de salariés). Les exercices (contexte : des commandes) sont tout en bas.

### Le df d'exemple

```python
import pandas as pd
df = pd.DataFrame({
    "nom": ["Alice", "Bob", "Chloé"],
    "salaire": [38000, 41000, 45000],
    "service": ["Data", "Data", "RH"],
})
df
```

### value_counts — compter les occurrences

```python
df["service"].value_counts()            # combien de fois chaque service apparaît
```

### isin — filtrer sur une liste de valeurs

```python
df[df["service"].isin(["Data", "RH"])]  # au lieu de combiner des | (ou)
```

### between — filtrer sur un intervalle (bornes incluses)

```python
df[df["salaire"].between(39000, 42000)] # salaires entre 39000 et 42000 inclus
```

### nlargest / nsmallest — top N direct

```python
df.nlargest(2, "salaire")               # les 2 plus gros salaires (tri + head en un appel)
```

### .str — méthodes texte sur toute la colonne

```python
print(df["nom"].str.upper().tolist())   # tout en majuscules
df[df["nom"].str.contains("o", case=False)]   # noms contenant "o" (casse ignorée)
```

### Enchaîner filtre -> tri -> sélection

```python
(
    df[df["service"] == "Data"]               # 1) filtre
      .sort_values("salaire", ascending=False) # 2) tri
      [["nom", "salaire"]]                     # 3) colonnes
)
```
