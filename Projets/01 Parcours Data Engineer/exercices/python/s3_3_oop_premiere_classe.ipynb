### La classe exemple : un compte bancaire

```python
class CompteBancaire:                       # class + Nom en PascalCase
    def __init__(self, titulaire, solde=0):  # constructeur : appelé à la création
        self.titulaire = titulaire           # on range les données dans l'objet (self)
        self.solde = solde                   # solde = 0 par défaut

    def deposer(self, montant):              # une action de l'objet
        self.solde = self.solde + montant

    def retirer(self, montant):
        if montant > self.solde:             # une méthode peut décider
            print("Retrait refusé : solde insuffisant")
            return
        self.solde = self.solde - montant
```

### On crée des instances et on les manipule

```python
a = CompteBancaire("Victor", 100)   # __init__ s'exécute ici
b = CompteBancaire("Sarah", 500)

a.deposer(50)      # -> a.solde passe de 100 à 150
a.retirer(1000)    # -> refusé
print(a.titulaire, ":", a.solde)   # Victor : 150
print(b.titulaire, ":", b.solde)   # Sarah : 500  (indépendant de a)
```

# Exercice — un panier d'achats

**Contexte.** Écris une classe qui représente le **panier** d'un client. À sa création, le panier est **vide** et connaît le **nom du client**. Il doit ensuite savoir :

1. **Ajouter un article** (nom de produit + prix). *Indice : à la création, prépare un attribut qui est une **liste vide**.*
2. Dire **combien d'articles** il contient.
3. Donner le **montant total** à payer.
4. **Appliquer une remise** en % sur le total et renvoyer le montant après remise (sans modifier les prix stockés).

**Test.** Crée un panier, ajoute-lui **3 articles**, affiche le nombre d'articles, le total, et le total après **10 %** de remise. Crée un **2e** panier (autre client, **1 article**) et vérifie que les deux sont **indépendants**.

*À toi de décider les attributs et les méthodes. Tu peux ajouter une méthode utile en plus (retirer un article, vider le panier) si tu veux.*

```python
# ton code : la classe Panier
class Panier:
    def __init__ (self, nom):
        self.nom = nom
        self.liste_articles = []
        self.liste_prix = []
    
    def AjouterArticle (self, nom, prix):
        print("L'utilisateur : ", self.nom, " a ajouté l'article ", nom, " d'une valeur de ", prix, "€ à votre panier")
        self.liste_articles.append(nom)
        self.liste_prix.append(prix)
    
    def NombreArticles(self):
        count = len(self.liste_articles)
        if count == 0:
            print("\nAucun article dans le panier.")
        else:
            print("\nVotre liste de course contient maintenant ", count, " articles.")
        self.count = count
    
    def Montant (self):
        montant_euros = sum(self.liste_prix)
        print("\nMontant total : ", sum(self.liste_prix), "€")
        self.montant_euros = montant_euros
        
    def Remise(self, remise):
        print("\nMontant après remise : ", (self.montant_euros - (self.montant_euros * (remise/100))), "€")
        
    def ViderPanier(self):
        self.liste_articles=[]
        self.liste_prix = []
        self.count = 0
        print("\nVous avez vidé  votre panier :\n    - Panier d'articles : ", self.liste_articles, "\n    - Liste des prix des articles : ", self.liste_prix, "\n    - Montant total : ", self.count, "€")
```

```python
# ton code : les tests (2 paniers indépendants)
v = Panier("Vichou")
v.AjouterArticle("Carte pokemon", 0.50)
v.AjouterArticle("3DS", 500)
v.AjouterArticle("BD", 5.60)

v.NombreArticles()
v.Montant()
v.Remise(10)
v.ViderPanier()
```
