```python
def anagrammes(mot1, mot2):
    mot1 = mot1.replace(" ", "").upper()
    mot2 = mot2.replace(" ", "").upper()
    
    if len(mot1) != len(mot2):
        print("Ces 2 mots ne sont pas des anagrammes car ils n'ont pas le même nombre de lettres")
        return 
    else:
        print(f"Ces 2 mots sont possiblements des anagrammes car ils contiennent tout les deux {len(mot1)} lettres :\n")
    
    liste_lettre_non_valide = []
    compteur = 1
    
    for lettre in mot1:
            if lettre in mot2:
                mot2 = mot2.replace(lettre, "", 1)
                print(f"{compteur}. Lettre identique dans les 2 mots, lettre ", lettre)
                compteur+=1
            else:
                liste_lettre_non_valide.append(lettre)
                
    
    if len(mot2) == 0:
        print("\nLes 2 mots sont des anagrammes")
    else:
        print("\nLes 2 mots ne sont pas des anagrammes")      
        print(f"Liste des lettres non valides : {liste_lettre_non_valide}")  
            
```

```python
anagrammes("banane", "cabane")
```
