## 2. Profilage : ce qui revient le plus

Quelles **classes thérapeutiques** sont les plus représentées dans les ventes ? Et quelles **régions** ? Y a-t-il un déséquilibre ?

```python
# ton code
count_classes = df["classe_therapeutique"].value_counts() # On peut remarquer que c'est les classes Antihypertenseur (18) Supplément (14) les autres sont à 9 ou moins
count_region = df["region"].value_counts(sort=True) # On peut remarquer que les régions Hauts-de-France (8) et Auvergne-Rhône-Alpes (10) sont les moins représentées les autres sont à minimum 14 apparitions

print("Sur les 110 lignes du dataset on remarque qu'il y a : ", sum(count_classes), " lignes avec une classe thérapeutique\n\n", count_classes, "\n\n")
print("Sur les 110 lignes du dataset on remarque qu'il y a : ", sum(count_region), " lignes avec une région\n\n", count_region)

# On a bien un déséquilibre
```

## 3. Construire un indicateur métier

Le nombre de boîtes ne dit pas tout : ce qui compte, c'est le **chiffre d'affaires** (CA = boîtes × prix unitaire). Crée cet indicateur par vente, puis donne le **CA total** du semestre.

```python
# ton code
df["chiffre_daffaire"] = df["boites_vendues"] * df["prix_unitaire"]
df_sans_nan = df.dropna() # Plus pour m'avancer pour la suite pour mieux comprendre
ca_total_avec_nan = sum(df["chiffre_daffaire"])
ca_total_sans_nan = sum(df_sans_nan["chiffre_daffaire"])

print("Le CA total du semestre (si on compte les NaN) est de : ", ca_total_avec_nan, "\n")
print("Le CA total du semestre (si on compte les NaN) est de : ", ca_total_sans_nan, "\n")
```

1. Quels sont les **5 médicaments** au plus gros chiffre d'affaires (sur une vente) ?

```python
# ton code
top_ventes = df.nlargest(5, "chiffre_daffaire").sort_values("chiffre_daffaire", ascending=False)
top_ventes
```

```python
# ton code
top_prix_u = df.nlargest(5, "prix_unitaire").sort_values("prix_unitaire", ascending=False)
top_prix_u
```

2. Quelle est la répartition **remboursable / non remboursable** des ventes ?

```python
df_non_remboursable = df[df["remboursable"].str.contains("Non")].sort_values("medicament", ascending=False)
liste_medicaments = df_non_remboursable["medicament"].unique().tolist()

print(liste_medicaments)
```

```python
df_remboursable = df[df["remboursable"].str.contains("Oui")].sort_values("prix_unitaire", ascending=False)
liste_medicaments_2 = df_remboursable["medicament"].unique().tolist()

print(liste_medicaments_2)
```

```python
répartition_remboursable = (df["remboursable"].str.upper() == "OUI").value_counts()
répartition_remboursable
```

3. Isole les ventes au CA élevé (à toi de fixer un seuil pertinent et de le justifier en commentaire).

```python
nb_ligne_valide = len(df_sans_nan)
ca_total = sum(df_sans_nan["chiffre_daffaire"])

moyenne_ca = ca_total / nb_ligne_valide
print("Moyenne du chiffre d'affaires : ", moyenne_ca)
```

```python
# On va prendre tout les CA au dessus de moyenne_ca
df_ca_eleve = df[df["chiffre_daffaire"] > moyenne_ca].sort_values("chiffre_daffaire")
df_ca_eleve
```
