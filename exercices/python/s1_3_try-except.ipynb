# Le Try/Except (Exercices)

```python
## 1
def division(a, b):
    try:
        d = a/b
    except ZeroDivisionError as z:
        print ("Division par zéro impossible")
    else: 
        print("La division est réussi et votre résultat est :")
        return d

print(division(10, 2),"\n")
print(division(10, 0))

```

```python
## 2
def convertir_entier(texte):
    try:
        a =int(texte)
    except ValueError as v:
        print ("Erreur : ", v)
    else:
        print("La conversionn est réussi :")
        return type(a), a 
    

print(convertir_entier("8"),"\n")
print(convertir_entier("bonjour"))
```

```python
## 3
produit = {"nom": "Clavier", "prix": 89.9}

try:
    stock = produit["stock"]
except KeyError as k:
    print("Erreur : stock non renseigné")
else:
    print("Le stock est parfaitement renseigné et le voici : ", stock)
```
