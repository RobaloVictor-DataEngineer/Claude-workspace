```python
import csv  # module standard de Python pour lire/écrire des CSV
```

```python
CHEMIN = "data/produits.csv"  # chemin du fichier à lire (relatif au script)
CHEMIN_VIDE = "data/inexistant.csv"
vide = []
```

```python
def lire_produits(chemin):
    """Ouvre le CSV et renvoie la liste des produits valides (1 dict par produit)."""
    print(" ------ FONCTION lire_produits ------\n")
    
    try:
        with open(chemin, encoding="utf-8") as f:
            open_file = open(chemin)
            lecteur = csv.DictReader(open_file)
            liste_produits = []
            count = 1
            for dictionnaire in lecteur:
                print("-- Ligne numéro", count, "--")
                try:
                    dictionnaire["prix"] = float(dictionnaire["prix"])  # teste ET convertit en une seule fois
                except ValueError:
                    print("ATTENTION :", repr(dictionnaire["prix"]), "-> conversion échouée, ligne ignorée\n")
                    count += 1
                else:
                    print("Prix converti :", type(dictionnaire["prix"]), dictionnaire["prix"], "\n")
                    liste_produits.append(dictionnaire)  # on garde le produit COMPLET
                    count += 1 
    except FileNotFoundError:
        print("Désolé, fichier introuvable au chemin :", chemin, "- réessayez\n\n")
        return []  # liste vide pour que afficher_resume tombe sur sa branche "vide"

    print("Liste finale des produits valides :\n -->", liste_produits, "\n\n")

    # 4) On renvoie la liste des produits valides
    return liste_produits

# def lire_produits_courte(chemin):
    """Ouvre le CSV et renvoie la liste des produits valides (1 dict par produit)."""
    print(" ------ FONCTION lire_produits ------\n")
    
    try:
        with open(chemin, encoding="utf-8") as f:
            open_file = open(chemin)
            lecteur = csv.DictReader(open_file)
            liste_produits = []
            count = 1
            for dictionnaire in lecteur:
                try:
                    dictionnaire["prix"] = float(dictionnaire["prix"])  # teste ET convertit en une seule fois
                except ValueError:
                    print("ATTENTION :", repr(dictionnaire["prix"]), "-> conversion échouée, ligne ignorée\n")
                    count += 1
                else:
                    liste_produits.append(dictionnaire)  # on garde le produit COMPLET
                    count += 1 
    except FileNotFoundError:
        print("Désolé, fichier introuvable au chemin :", chemin, "- réessayez\n\n")
        return []  # liste vide pour que afficher_resume tombe sur sa branche "vide"

    print("Liste finale des produits valides :\n -->", liste_produits, "\n\n")

    # 4) On renvoie la liste des produits valides
    return liste_produits

def afficher_resume(produits):
    """Affiche un résumé : nombre de produits valides + prix moyen."""

    # 5) Liste vide -> on le signale et on s'arrête proprement
    if len(produits) == 0:
        print("------ afficher_resume (liste vide) ------\n")
        print("Liste vide, rien à résumer.\n\n")
        return

    print("------ afficher_resume (liste non vide) ------\n")

    # 6) Combien de produits chargés
    print("Nombre de produits chargés :", len(produits))

    # 7) Prix moyen = somme des prix / nombre de produits
    somme_prix = 0
    for produit in produits:
        somme_prix = somme_prix + produit["prix"]

    print("Prix moyen :", round(somme_prix / len(produits), 2), "€")
```

```python
# Point d'entrée du script
if __name__ == "__main__":  # ne s'exécute que si on lance CE fichier directement
    produits_vide = lire_produits(CHEMIN_VIDE)
    produits = lire_produits(CHEMIN)
    afficher_resume(produits_vide)
    afficher_resume(produits)
```
